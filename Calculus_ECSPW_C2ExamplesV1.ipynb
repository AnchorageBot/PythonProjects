Let's review some rules of algebra

* 1/$x^n$ = $x^{-n}$
* $\sqrt{x}$ = x$^{1/2}$
* $x^0$ = 1
* $x^m$$x^n$ = $x^{m+n}$
* $x^m$/$x^n$ = $x^{m-n}$
* ($x^m$)$^n$ = $x^{mn}$
* (ax)$^m$ = $a^m$$x^m$
*  $\sqrt{ax}$ = (ax)$^{1/2}$ = $a^{1/2}$$x^{1/2}$
* A function assigns each element of a set (a mathematical collection of things) X to exactly one element of a set Y.

Remember that a derivative:
* Describes the rate of change
* The operation itself can denoted by a variety of symbols
> * f'(t) = df/dt = d/dt {f(t)}
* Can be solved by selectively applying rules
> * d/dx (constant c) = 0
> * d/dx (a$x^b$) = bax$^{b-1}$
> * d/dx ($y_1$ + $y_2$ + ... + $y_n$) = d$y_1$/dx + d$y_2$/dx + ... + d$y_n$/dx
> * Chain rule
> > * Finds the derivative of f(u(x)) with respect to x
> > * f ' (u(x)) = $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$
> * Product rule
> > * Finds the derivative of the product of f(x) and g(x) with respect to x
> > * f ' (fg) = $\frac{d}{dx}$ (fg) = g($\frac{df}{dx}$) + f($\frac{dg}{dx}$)
> * Quotient rule
> > * Finds the derivative of the quotient of f(x) and g(x) with respect to x
> > * f ' ($\frac{f}{g}$) = $\frac{d}{dx}$ ($\frac{f}{g}$) = [g($\frac{df}{dx}$) - f($\frac{df}{dx}$)]/g$^2$

Let's take the derivative of f(u(x)) = (2$x^2$ - 5)$^5$
* We will use the chain rule
> * f ' (u(x)) = $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$
* This gives us
> * f(u(x)) = (2$x^2$ - 5)$^5$
> * f(u) = u$^5$
> * u(x) = 2x$^2$ - 5
> * f ' (u(x)) = $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$
> * * $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$ = $\frac{d}{du}$(u^$5$) $\frac{d}{dx}$ (2x$^2$ - 5)
> * * $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$ = (5u$^4$)(4x)
> * * $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$ = 20xu$^4$
> * u = 2x$^2$ - 5
> * f ' (u(x)) = 20x(2x$^2$ - 5)$^4$

Let's graph the function of f(u(x)) = (2$x^2$ - 5)$^5$ with matplotlib:

```python
import matplotlib.pyplot as plt

xmin = -4
xmax = 4
ymin = -4000
ymax = 4000

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.title('function f(u(x)) = (2x^2 - 5)^5')
plt.xlabel('x')
plt.ylabel('y = (2x^2 - 5)^5')

for x in range(-5,5): # this graphs a line up and to the left and to the right
    y = (((2*x**2)-5)**5) # rate of change
    plt.plot([x],[y], 'ro') # r = red , o = marker, this makes the red dots



plt.show()
```

Let's calculate the derivative of (2x$^2$ -5)$^5$ with respect to x, and check our answer by integrating it, using SymPy. Then let's graph it using matplotlib

```python
import matplotlib.pyplot as plt

from sympy import *
x = symbols('x')
#init_printing(use_unicode = True)

x_deriv = diff(((2*x**2)-5)**5,x)
print("The derivative of ((2*x**2)-5)**5 with respect to x is", x_deriv, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w = integrate(x_deriv)
print("The integral of 20*x*(2*x**2 - 5)**4 is", w, "\n")
print("\n")

print('The plot of the derivative \n')
plot(x_deriv)
print("\n")

print('The plot of the integral \n')
plot(w)
print("\n")
```

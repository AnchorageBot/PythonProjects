## Equations and Derivations

### Step 1: Start with the Weierstrass Product Formula
The Weierstrass factorization theorem for \( \sin(\pi x) \) is given as:

1. $$ \sin(\pi x) = \pi x \prod_{k=1}^\infty \left( 1 - \frac{x^2}{k^2} \right) $$

This expresses \( \sin(\pi x) \) as an infinite product involving \( x \) and the roots of the sine function.

---

### Step 2: Take the Natural Logarithm of Both Sides
By applying the natural logarithm to both sides, we obtain:

2. $$ \ln(\sin(\pi x)) = \ln(\pi x) + \sum_{k=1}^\infty \ln\left( 1 - \frac{x^2}{k^2} \right) $$

Here, the logarithm simplifies the product into a summation, which is easier to differentiate.

---

### Step 3: Differentiate the Logarithm
Differentiating the logarithmic expression with respect to \( x \), we get:

3. $$ \frac{\pi \cos(\pi x)}{\sin(\pi x)} = \frac{1}{x} + \sum_{k=1}^\infty \left( \frac{1}{1 - \frac{x^2}{k^2}} \cdot \frac{-2x}{k^2} \right) $$

Simplify the summation term:

$$ = \frac{1}{x} - 2x \sum_{k=1}^\infty \frac{1}{k^2 - x^2} $$

This uses the chain rule and the fact that \( \frac{d}{dx} \ln(1 - u) = \frac{-1}{1 - u} \cdot \frac{du}{dx} \).

---

### Step 4: Rearrange to Isolate the Summation
Rearranging the above equation to focus on the summation:

4. $$ \frac{\pi \cos(\pi x)}{\sin(\pi x)} - \frac{1}{x} = -2x \sum_{k=1}^\infty \frac{1}{k^2 - x^2} $$

---

### Step 5: Solve for the Infinite Series
Divide through by \( -2x \) to isolate the series:

5. $$ \left( \frac{\pi \cos(\pi x)}{\sin(\pi x)} - \frac{1}{x} \right) \cdot \frac{-1}{2x} = \sum_{k=1}^\infty \frac{1}{k^2 - x^2} $$

---

### Step 6: Analyze the Expression as \( x \to 0 \)
By expanding the sine function using its Taylor series, we find that:

6. $$ \frac{\sin(\pi x) - \pi x \cos(\pi x)}{2x^2 \sin(\pi x)} = \sum_{k=1}^\infty \frac{1}{k^2 - x^2}, \quad k^2 - x^2 \neq 0 $$

---

### Step 7: Address the Indeterminate Form at \( x = 0 \)
If \( x = 0 \), the left-hand side becomes:

7. $$ \sum_{k=1}^\infty \frac{1}{k^2} = 1 + \frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \cdots $$

This is the Basel Problem series.

The left-hand side leads to an indeterminate form:

8. $$ \lim_{x \to 0} \frac{\sin(\pi x) - \pi x \cos(\pi x)}{2x^2 \sin(\pi x)} \to \frac{0}{0} $$

---

### Step 8: Resolve Using L’Hôpital’s Rule
Apply L’Hôpital’s Rule repeatedly to evaluate the limit:

#### First Application:
9. $$ \lim_{x \to 0} \frac{\sin(\pi x) - \pi x \cos(\pi x)}{2x^2 \sin(\pi x)} = \lim_{x \to 0} \frac{\pi^2 x \sin(\pi x)}{4x \sin(\pi x) + 2\pi x^2 \cos(\pi x)} $$

Still indeterminate (\( \frac{0}{0} \)).

#### Second Application:
10. $$ \lim_{x \to 0} \frac{\pi^3 \cos(\pi x) x + \pi^2 \sin(\pi x)}{4\sin(\pi x) + 8\pi \cos(\pi x)x - 2\pi x^2 \sin(\pi x)} $$

Still indeterminate (\( \frac{0}{0} \)).

#### Third Application:
11. $$ \lim_{x \to 0} \frac{2\pi^3 \cos(\pi x) - \pi^4 \sin(\pi x)}{12\pi \cos(\pi x)} $$

---

### Step 9: Evaluate the Limit
At \( x = 0 \), simplify to:

12. $$ \frac{2\pi^3}{12\pi} = \frac{\pi^2}{6} $$

---

### Step 10: Conclude the Proof
Thus, the Basel Problem sum is:

13. $$ \sum_{k=1}^\infty \frac{1}{k^2} = \frac{\pi^2}{6} $$

This completes the proof of Euler’s identity.

```python
# Made in MATLAB by ATS
# Converted to Python by Anthropic Claude

import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
RGB = cv2.imread('HSV1.jpg')

# Get the size (rows and columns) of the image
r, c, _ = RGB.shape
print(f"r={r}, c={c}")

# Show image
plt.imshow(cv2.cvtColor(RGB, cv2.COLOR_BGR2RGB))
plt.show()

# Get the size (rows and columns) of the subimage
rr = r // 2
cc = c // 6

# Get a subimage
subimg1 = RGB[0:400, 0:655]

# Show subimage
plt.imshow(cv2.cvtColor(subimg1, cv2.COLOR_BGR2RGB))
plt.show()

# Montage Function displays multiple image frames as rectangular montage
titles = ['RGB', 'HSV', 'NTSC', 'YCbCr']
images = [RGB]

# Convert RGB color values to HSV color space
HSV = cv2.cvtColor(RGB, cv2.COLOR_BGR2HSV)
images.append(HSV)

# Convert RGB color values to YCrCb color space
YCrCb = cv2.cvtColor(RGB, cv2.COLOR_BGR2YCrCb)
images.append(YCrCb)

# Convert RGB color values to YCbCr color space
YCbCr = cv2.cvtColor(RGB, cv2.COLOR_BGR2YCbCr)
images.append(YCbCr)

# Montage the fusion
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for i, ax in enumerate(axs.flat):
    ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    ax.set_title(titles[i])
    ax.axis('off')
plt.show()
```

```python
"""
Abstract:
This script generates a Fibonacci spiral and plots the "arms" or parastichies
commonly observed in sunflower seed heads.  It uses polar coordinates to
represent the spiral and then converts them to Cartesian coordinates for
plotting. The script highlights the mathematical relationship between
the Fibonacci sequence and the spiral patterns found in nature. It focuses
on visualizing the arrangement of the "arms" rather than the individual
seeds for clarity.

Author: Gemini (AI Assistant)

Date: October 26, 2023

Today's date is Jan 27, 2025
"""

import matplotlib.pyplot as plt
import numpy as np

def fibonacci_spiral(n_spirals=1000, scaling_factor=0.5):
    """
    Generates coordinates for a Fibonacci spiral.

    Args:
        n_spirals (int): The number of points/spirals to generate.
        scaling_factor (float): Scales the spiral's growth.

    Returns:
        tuple: Two numpy arrays representing the x and y coordinates of the spiral.
    """

    # Golden angle (approximately 137.5 degrees).  This angle is crucial for the
    # formation of the spiral patterns seen in nature.
    golden_angle = 137.508  # In degrees

    # Convert the angle to radians for trigonometric calculations.
    golden_angle_rad = np.radians(golden_angle)

    # Generate a sequence of numbers from 0 to n_spirals-1.  These will be used
    # to calculate the radius and angle of each point on the spiral.
    n = np.arange(n_spirals)

    # Calculate the radius of each point.  The radius increases with the square
    # root of the index, creating the spiral effect.  The scaling factor
    # controls how tightly or loosely the spiral winds.
    radius = scaling_factor * np.sqrt(n)

    # Calculate the angle of each point.  We multiply the index by the golden
    # angle (in radians) to distribute the points along the spiral.
    angle = n * golden_angle_rad

    # Convert polar coordinates (radius, angle) to Cartesian coordinates (x, y).
    x = radius * np.cos(angle)
    y = radius * np.sin(angle)

    return x, y


def plot_fibonacci_spiral(x, y, title="Fibonacci Spiral"):
    """
    Plots the Fibonacci spiral.

    Args:
        x (numpy.ndarray): The x-coordinates of the spiral.
        y (numpy.ndarray): The y-coordinates of the spiral.
        title (str): The title of the plot.
    """

    plt.figure(figsize=(8, 8))  # Set the figure size for better visualization.
    plt.scatter(x, y, s=1, c='blue', marker='o') # Plot the spiral points. s controls size
                                                  # c controls colour
                                                  # marker controls point style
    plt.title(title)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.axis('equal')  # Ensure that the x and y axes are scaled equally,
                        # preventing distortion of the spiral.
    plt.grid(False)    # Turn off the grid for a cleaner look.
    plt.xticks([])     # Remove x-axis ticks for a cleaner look.
    plt.yticks([])     # Remove y-axis ticks for a cleaner look.
    plt.show()




if __name__ == "__main__":
    x_coords, y_coords = fibonacci_spiral(n_spirals=1000, scaling_factor=1)
    plot_fibonacci_spiral(x_coords, y_coords, title="Fibonacci Spiral (Sunflower Pattern)")



    # Demonstrating different scaling factors.  You can uncomment these to
    # see how the spiral changes.

    # x_coords, y_coords = fibonacci_spiral(n_spirals=1000, scaling_factor=0.5)
    # plot_fibonacci_spiral(x_coords, y_coords, title="Fibonacci Spiral (Smaller Scaling)")

    # x_coords, y_coords = fibonacci_spiral(n_spirals=1000, scaling_factor=2)
    # plot_fibonacci_spiral(x_coords, y_coords, title="Fibonacci Spiral (Larger Scaling)")

    # Demonstrating different numbers of spirals

    # x_coords, y_coords = fibonacci_spiral(n_spirals=500, scaling_factor=1)
    # plot_fibonacci_spiral(x_coords, y_coords, title="Fibonacci Spiral (Fewer Spirals)")

    # x_coords, y_coords = fibonacci_spiral(n_spirals=2000, scaling_factor=1)
    # plot_fibonacci_spiral(x_coords, y_coords, title="Fibonacci Spiral (More Spirals)")
```

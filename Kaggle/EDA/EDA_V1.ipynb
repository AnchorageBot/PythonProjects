Our first script loads the following libraries:
*  [Pandas](https://pandas.pydata.org/) version 2.14, dated 8 December 2023
*  [Seaborn](https://seaborn.pydata.org/) version v 0.13.1, dated 2021
*  [Matplotlib](https://matplotlib.org/) version 3.8.2, dated 17 Nov, 2023

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Our next script:
*  Uses the Pandas library
*  Loads the data
*  Uses the [.columns method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.columns.html)
*  Displays the column labels in the dataset

```python
# Load data
home_data = pd.read_csv('train.csv')

# Examine the column labels in the dataset
home_data.columns
```

This script:
*  Uses the Pandas library
*  Uses the [.info method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html)
*  Provides a concise summary of the dataset

```python
home_data.info()
```

Our next script:
*  Uses the Seaborn library
*  Uses the [.heatmap method](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
*  Plots rectangular data as a color-encoded matrix
*  Uses the Panda library
*  Uses the [.corr method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html)
*  Computes pairwise correlation of columns, excluding NA/null values
*  Uses the matplotlib library
*  Uses the [.rcParams method](https://matplotlib.org/stable/users/explain/customizing.html)
*  You can dynamically change the default rc (runtime configuration) settings in a python script

```python
sns.heatmap(home_data.corr(), annot = False)

plt.rcParams['figure.figsize'] = (10,9)

plt.show()
```

Our next script:

*  Uses the Seaborn library
*  Uses the [.boxplot method](https://seaborn.pydata.org/generated/seaborn.boxplot.html)
*  Draw a box plot (box and whisker plot) to show distributions with respect to categories
*  Uses the [.set_xticklabels method](https://seaborn.pydata.org/generated/seaborn.FacetGrid.set_xticklabels.html)
*  Sets the angle of x axis tick labels of the grid



```python
# create box plot (box and whisker plot)
whisker = home_data.boxplot(
    figsize=(15,7)
    )

# rotate x-axis labels
whisker.set_xticklabels(whisker.get_xticklabels(), rotation=45)
```

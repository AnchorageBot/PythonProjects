## Given:

* h(x) = $ \frac{4x}{(x^2 + 4)^{0.5}} $

* $ \frac{d}{dx} (ax^b) = bax^{b-1} $

* $ \frac{d}{dx} f(u(x)) = \frac{df}{du} \cdot \frac{du}{dx} $

* $ \frac{d}{dx} \left(\frac{f}{g}\right) = \frac{g \cdot \frac{df}{dx} - f \cdot \frac{dg}{dx}}{g^2} $


##Find:

$
\frac{d}{dx} h(x)
$

# Solution
h(x) = x⁴/√(x² + 4)

1) Identify composite functions & differentiate
* d/dx (ax^b) = bax^(b-1)
* v(x) = x² + 4
* u = v^(1/2)
* dv/dx = v'(x) = 2x
* du/dx = u'(x) = (1/2)·v^(-1/2)

2) Apply the chain rule
* d/dx (v(u)) = (dv/du)·(du/dx)
* dv/dx = (dv/du)·(du/dx) = (2x)·(1/2·v^(-1/2)) = (2x)·(1/2)·(1/√(x²+4)) = x/√(x²+4)
* f(x) = x⁴
* g(x) = √(x²+4)
* f'(x) = 4x³
* g'(x) = x/√(x²+4) = (x·√(x²+4))/(x²+4)

3) Apply the quotient rule
* d/dx (f/g) = (g·df/dx - f·dg/dx)/g²
* d/dx (f/g) = (√(x²+4)·(4x³) - (x⁴)·(x·√(x²+4)/(x²+4)))/(√(x²+4))²
* = (4x³√(x²+4) - (x⁵)(√(x²+4))/(x²+4))/(x²+4)
* = (4x³√(x²+4) - x⁵/√(x²+4))/(x²+4)
* = 4x³√(x²+4)/(x²+4) - x⁵(x²+4)^(-1/2)/(x²+4)
* = 4x³(x²+4)^(-1/2) - x⁵(x²+4)^(-3/2)

4) Simplify answwer
* d/dx (f/g) = 4x^3 (x^2+4)^(-1/2) - x^5 (x^2+4)^(-3/2)

= 4x^3/(x^2+4)^(1/2) - x^5/(x^2+4)^(3/2)

= 4x^3/(x^2+4)^(1/2) * (x^2+4)^(2/2)/(x^2+4)^(2/2) - x^5/(x^2+4)^(3/2)

= 4x^3 (x^2+4)^(2/2)/(x^2+4)^(3/2) - x^5/(x^2+4)^(3/2)

= [4x^3 (x^2+4) - x^5]/(x^2+4)^(3/2)

= [4x^5-x^5 + 16x^3]/(x^2+4)^(3/2)

= [3x^5 + 16x^3]/(x^2+4)^(3/2)

Wolfram Alpha

d/dx(x^4/sqrt(x^2 + 4)) = (x^3 (16 + 3 x^2))/(4 + x^2)^(3/2)

```python
# Script made by Google Gemini

import sympy as sp

# Define the variable and the function
x = sp.Symbol('x')
h = (x**4) / sp.sqrt(x**2 + 4)

# Calculate the derivative
h_prime = sp.diff(h, x)

# Simplify the derivative
h_prime_simplified = sp.simplify(h_prime)

# Display the original function and its derivative
print("Original function h(x):", h)
print("Derivative of h(x):", h_prime_simplified)

# Compare with the result in the notebook: (3*x**5 + 16*x**3) / (x**2 + 4)**(3/2)
notebook_result = (3*x**5 + 16*x**3) / (x**2 + 4)**(3/2)

# Check if the derivative matches the result in the notebook
print("Derivative matches notebook result:", h_prime_simplified == notebook_result)
```

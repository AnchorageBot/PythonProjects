```python
# This script creates a simple animation of a square transforming into a circle

from manim import *

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        self.play(Create(square))
        self.play(Transform(square, circle))
        self.play(FadeOut(square))

# This part is specific to Colab
config.media_width = "60%"
config.verbosity = "WARNING"
#%manim -v WARNING -qm SquareToCircle
#%%manim -v WARNING -qm SquareToCircle
#%%manim -v WARNING -qm --format=gif SquareToCircle
%manim -v WARNING -qm --format=gif SquareToCircle
from IPython.display import Image
Image('media/videos/1080p60/SquareToCircle.gif')
```

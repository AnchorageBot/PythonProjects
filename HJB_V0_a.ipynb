## Overview
A Hamiltonian is a mathematical function used in classical mechanics to describe the state of a physical system. It represents the total energy of the system and is fundamental in formulating equations of motion and in the study of conserved quantities.

## Key Concepts

1. **Definition**: The Hamiltonian H is typically defined as the sum of kinetic energy T and potential energy V:
   
   H = T + V

2. **Variables**: Expressed in terms of generalized coordinates (q) and momenta (p).

3. **Hamilton's Equations**: Describe the evolution of a system:
   - dq/dt = ∂H/∂p
   - dp/dt = -∂H/∂q

4. **Conservation**: In many systems, the Hamiltonian is conserved (constant over time).

5. **Phase Space**: The Hamiltonian defines the geometry of the phase space, a 2n-dimensional space where n is the number of degrees of freedom.

6. **Relation to Lagrangian**: Obtained from the Lagrangian L via Legendre transformation:
   
   H(q, p, t) = Σ(p_i * dq_i/dt) - L(q, dq/dt, t)

7. **Applications**:
   - Classical Mechanics
   - Quantum Mechanics (operator form)
   - Statistical Mechanics
   - Control Theory

8. **Importance in Physics**:
   - Simplifies many physical problems
   - Basis for Hamilton-Jacobi theory
   - Fundamental in quantum mechanics and statistical physics

9. **Extensions**: Generalized to fields (e.g., electromagnetic fields) and relativistic systems.

## Glossary of Terms and Math Symbols

| Term/Symbol | Definition |
|-------------|------------|
| H           | Hamiltonian function |

| V           | Potential energy |

| p_i         | i-th component of momentum |
| q_i         | i-th component of position |

```python
# Discrete Hamiltonian Simulation: Simple Harmonic Oscillator

## Abstract

'''
This script simulates a simple harmonic oscillator using Hamilton's equations. It demonstrates
the conservation of energy (Hamiltonian) in a discrete-time simulation. The system consists of a mass
on a spring, where the Hamiltonian is the sum of kinetic and potential energies. The script uses
a basic Euler method for numerical integration and visualizes the position, momentum, and total energy over time.
'''

## Python Implementation

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# System parameters
m = 1.0  # mass
k = 1.0  # spring constant
dt = 0.1  # time step
t_max = 20  # total simulation time

# Initial conditions
q0 = 1.0  # initial position
p0 = 0.0  # initial momentum

# Arrays to store results
t = np.arange(0, t_max, dt)
q = np.zeros_like(t)
p = np.zeros_like(t)
H = np.zeros_like(t)

# Set initial conditions
q[0] = q0
p[0] = p0

# Hamiltonian function
def hamiltonian(q, p):
    return 0.5 * p**2 / m + 0.5 * k * q**2

# Hamilton's equations
def dqdt(p):
    return p / m

def dpdt(q):
    return -k * q

# Simulation loop
for i in range(1, len(t)):
    # Update position using previous momentum
    q[i] = q[i-1] + dqdt(p[i-1]) * dt

    # Update momentum using new position
    p[i] = p[i-1] + dpdt(q[i]) * dt

    # Calculate and store Hamiltonian (total energy)
    H[i] = hamiltonian(q[i], p[i])

# Calculate initial Hamiltonian
H[0] = hamiltonian(q[0], p[0])

# Plotting
plt.figure(figsize=(12, 8))

# Plot position
plt.subplot(3, 1, 1)
plt.plot(t, q)
plt.title('Simple Harmonic Oscillator Simulation')
plt.ylabel('Position (q)')

# Plot momentum
plt.subplot(3, 1, 2)
plt.plot(t, p)
plt.ylabel('Momentum (p)')

# Plot total energy (Hamiltonian)
plt.subplot(3, 1, 3)
plt.plot(t, H)
plt.ylabel('Total Energy (H)')
plt.xlabel('Time')

# Adjust layout and display
plt.tight_layout()
plt.show()

# Print final energy and compare with initial
print(f"Initial energy: {H[0]:.6f}")
print(f"Final energy: {H[-1]:.6f}")
print(f"Energy difference: {abs(H[-1] - H[0]):.6f}")
```

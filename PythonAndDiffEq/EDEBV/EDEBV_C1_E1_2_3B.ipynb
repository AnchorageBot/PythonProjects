Given:
* 2nd order ODE y" + 2y' + y = 2x
* Constants $c_1 and c_2$
* $y = (c_1 + c_2x)e^{-x} + 2x - 4$

Find:
* Show that $y = (c_1 + c_2x)e^{-x} + 2x - 4$ is a solution of y" + 2y' + y = 2x on $(-{\infty}, {\infty})$


Solution: (Strategy)

* Differentiate y to get the resulting y' term
  
  * Use the constant term rule $\frac{d}{dx}(c) = 0$

  * Use the power rule $\frac{d}{dx}(x^n) = nx^{n-1}$

  * Use the chain rule $\frac{d h(x)}{dx} = \frac{df(g(x))}{dx} \cdot \frac{d g(x)}{dx}$

* Differentiate y' to get the resulting y" term

* Substitute the y' and y" terms back into y" + 2y' + y = 2x

* Simplify

Solution: (Step by step)

1. Given equation: y" + 2y' + y = 2x

   Given solution: y = (c1 + c2x)e^(-x) + 2x - 4

2. Let's differentiate y to get y':

   y' = [(c1 + c2x)(-1)e^(-x) + c2e^(-x)] + 2
      = -c1e^(-x) - c2xe^(-x) + c2e^(-x) + 2
      = (-c1 - c2x + c2)e^(-x) + 2

3. Now let's differentiate y' to get y":

   y" = [(-c1 - c2x + c2)(-1)e^(-x) - c2e^(-x)]
      = (c1 + c2x - c2)e^(-x) - c2e^(-x)
      = (c1 + c2x - 2c2)e^(-x)

4. Now we substitute y, y', and y" into the original equation:

   y" + 2y' + y = 2x

   [(c1 + c2x - 2c2)e^(-x)] + 2[(-c1 - c2x + c2)e^(-x) + 2] + [(c1 + c2x)e^(-x) + 2x - 4] = 2x

5. Let's simplify the left hand side (LHS):

   (c1 + c2x - 2c2)e^(-x) + (-2c1 - 2c2x + 2c2)e^(-x) + 4 + (c1 + c2x)e^(-x) + 2x - 4 = 2x

   Combining like terms:
   (c1 + c2x - 2c2 - 2c1 - 2c2x + 2c2 + c1 + c2x)e^(-x) + 2x = 2x

   Simplifying inside the parentheses:
   (0)e^(-x) + 2x = 2x

   2x = 2x

This proves that the given function y = (c1 + c2x)e^(-x) + 2x - 4 is indeed a solution to the differential equation y" + 2y' + y = 2x for all x in (-∞, ∞).


```python
"""
Abstract:
This script visualizes the equation y = (c1 + c2x)e^(-x) + 2x - 4, where c1 and c2 are constants.
It demonstrates the use of numpy for numerical computations and matplotlib for creating a
professional-looking plot. The script allows for easy modification of the constants c1 and c2,
enabling users to explore how these parameters affect the shape of the function.

The plot includes proper labeling, a title with LaTeX formatting, a legend, grid lines,
and axes at y=0 and x=0 for reference. This visualization aids in understanding the behavior
of the given differential equation solution across a range of x values.
"""

import numpy as np
import matplotlib.pyplot as plt

def y(x, c1, c2):
    """
    Calculate y values for the equation y = (c1 + c2x)e^(-x) + 2x - 4.

    Args:
    x (array-like): Input x values
    c1 (float): First constant in the equation
    c2 (float): Second constant in the equation

    Returns:
    array-like: Calculated y values
    """
    return (c1 + c2*x) * np.exp(-x) + 2*x - 4

# Generate x values
# We use numpy's linspace to create 1000 evenly spaced points from -5 to 5
x = np.linspace(-5, 5, 1000)

# Set values for c1 and c2
# These can be modified to explore different variations of the function
c1, c2 = 1, 2

# Calculate y values using our defined function
y_values = y(x, c1, c2)

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size to 10x6 inches

# Plot the function
plt.plot(x, y_values, label=f'c1={c1}, c2={c2}')

# Set the title with LaTeX formatting
plt.title(r'Plot of $y = (c_1 + c_2x)e^{-x} + 2x - 4$')

# Label the axes
plt.xlabel('x')
plt.ylabel('y')

# Add a horizontal line at y=0
plt.axhline(y=0, color='k', linestyle='--', linewidth=0.5)

# Add a vertical line at x=0
plt.axvline(x=0, color='k', linestyle='--', linewidth=0.5)

# Add a grid to the plot
plt.grid(True, linestyle=':', alpha=0.7)

# Add a legend to identify the plotted function
plt.legend()

# Display the plot
plt.show()
```

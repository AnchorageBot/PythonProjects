Probability Density Functions (PDFs):

A probability density function is a function that describes the relative likelihood for a continuous random variable to take on a given value. The integral of a PDF over an interval gives the probability that the variable falls within that interval. Key properties of PDFs include:

1. They are always non-negative.
2. The total area under the curve must equal 1.
3. They are used for continuous random variables, as opposed to probability mass functions for discrete variables.

Now, let's create a Python script with three examples of PDFs: Normal (Gaussian) distribution, Exponential distribution, and Uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set up the plot
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
x = np.linspace(-5, 5, 1000)

# 1. Normal (Gaussian) Distribution
mu, sigma = 0, 1
y_normal = stats.norm.pdf(x, mu, sigma)
axs[0].plot(x, y_normal)
axs[0].set_title('Normal Distribution (μ=0, σ=1)')
axs[0].set_ylabel('Probability Density')

# 2. Exponential Distribution
lambda_param = 1
y_exp = stats.expon.pdf(x, scale=1/lambda_param)
axs[1].plot(x, y_exp)
axs[1].set_title(f'Exponential Distribution (λ={lambda_param})')
axs[1].set_ylabel('Probability Density')

# 3. Uniform Distribution
a, b = -2, 2
y_uniform = stats.uniform.pdf(x, loc=a, scale=b-a)
axs[2].plot(x, y_uniform)
axs[2].set_title(f'Uniform Distribution (a={a}, b={b})')
axs[2].set_ylabel('Probability Density')
axs[2].set_xlabel('x')

plt.tight_layout()
plt.show()
```

This script creates three subplots, each showing a different probability density function:

1. Normal (Gaussian) Distribution: A symmetric bell-shaped curve centered at the mean (μ=0) with a standard deviation (σ=1).

2. Exponential Distribution: Often used to model the time between events in a Poisson process, with rate parameter λ=1.

3. Uniform Distribution: Represents equal probability over a fixed range, in this case from a=-2 to b=2.

Each plot shows the shape of the PDF, illustrating how the probability density varies across different values of the random variable.

````python
'''
Abstract:
This Python script demonstrates three common probability density functions (PDFs):
Normal (Gaussian), Exponential, and Uniform distributions. It uses NumPy for numerical
operations, Matplotlib for plotting, and SciPy for statistical functions. The script
creates a figure with three subplots, each displaying a different PDF. This visualization
helps in understanding the shape and characteristics of these fundamental probability
distributions, which are crucial in various fields of statistics, data science, and engineering.
'''
# Import necessary libraries
import numpy as np  # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for plotting
from scipy import stats  # SciPy stats module for statistical functions

# Set up the plot: Create a figure with 3 subplots arranged vertically
# figsize=(10, 15) sets the overall figure size to 10 inches wide and 15 inches tall
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Create an array of 1000 equally spaced points from -5 to 5
# This will be our x-axis for all three plots
x = np.linspace(-5, 5, 1000)

# 1. Normal (Gaussian) Distribution
mu, sigma = 0, 1  # Set mean (mu) to 0 and standard deviation (sigma) to 1
y_normal = stats.norm.pdf(x, mu, sigma)  # Calculate the PDF values for normal distribution
axs[0].plot(x, y_normal)  # Plot the normal distribution on the first subplot
axs[0].set_title('Normal Distribution (μ=0, σ=1)')  # Set the title for this subplot
axs[0].set_ylabel('Probability Density')  # Label the y-axis

# 2. Exponential Distribution
lambda_param = 1  # Set the rate parameter lambda to 1
# Calculate the PDF values for exponential distribution
# Note: scale=1/lambda_param because SciPy uses scale parameter which is 1/lambda
y_exp = stats.expon.pdf(x, scale=1/lambda_param)
axs[1].plot(x, y_exp)  # Plot the exponential distribution on the second subplot
axs[1].set_title(f'Exponential Distribution (λ={lambda_param})')  # Set the title
axs[1].set_ylabel('Probability Density')  # Label the y-axis

# 3. Uniform Distribution
a, b = -2, 2  # Set the lower bound 'a' to -2 and upper bound 'b' to 2
# Calculate the PDF values for uniform distribution
# loc is the lower bound, scale is the range (upper - lower)
y_uniform = stats.uniform.pdf(x, loc=a, scale=b-a)
axs[2].plot(x, y_uniform)  # Plot the uniform distribution on the third subplot
axs[2].set_title(f'Uniform Distribution (a={a}, b={b})')  # Set the title
axs[2].set_ylabel('Probability Density')  # Label the y-axis
axs[2].set_xlabel('x')  # Label the x-axis (only needed for the bottom subplot)

# Adjust the layout to prevent overlapping
plt.tight_layout()

# Display the plot
plt.show()
````

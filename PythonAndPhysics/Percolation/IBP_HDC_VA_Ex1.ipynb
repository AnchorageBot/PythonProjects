Exercise 1
- - - -

Show that any measurable event can be approximated by events depending on finitely many edges, in
the sense that for any A in the product σ-algebra, there exists a sequence (Bn) such that Bn is measurable in terms
of the ωe for e in the box of size n, and Pp[A∆Bn] tends to 0 as n tends to infinity.

## **Reformatted Exercise 1**

### **Given**  
- **𝐺 = (𝑉, 𝐸)**: A graph consisting of a set of vertices **𝑉** and edges **𝐸**.  
- **𝑉**: The set of vertices (nodes) in the graph.  
- **𝐸**: The set of edges (connections) between vertices.  
- **𝜎-algebra (𝓕)**: The collection of all events measurable in terms of a percolation configuration.  
- **𝑒**: An edge in the graph, typically denoted as an unordered pair of vertices {𝑥, 𝑦}.  
- **𝜔ₑ = 1**: The edge **𝑒** is open in the percolation configuration.  
- **𝜔ₑ = 0**: The edge **𝑒** is closed in the percolation configuration.  
- **𝑃ₚ[𝐴]**: The probability of an event **𝐴** occurring in Bernoulli percolation with parameter **𝑝**.  
- **𝒫(𝐵ₙ)**: The probability that an event **𝐵ₙ** is measurable in terms of a percolation configuration.  

---

### **Find**  
Show that **any measurable event can be approximated by events depending on finitely many edges**.  
That is, for any **𝐴** in the product **𝜎-algebra 𝓕**, there exists a sequence of events **(𝐵ₙ)** such that:
\[
\mathbb{P}_p(𝐵ₙ) \text{ is measurable in terms of } 𝜔_e
\]
for **finitely many edges**, and  
\[
𝒫ₚ[𝐵ₙ] \to 𝒫ₚ[𝐴] \text{ as } n \to \infty.
\]

---

### **Solution (Approach Outline)**  
1. **Interpretation of Measurability in Terms of Edges**  
   - The probability measure **𝒫ₚ** is defined on the **𝜎-algebra 𝓕**, which is generated by events that depend only on a finite subset of edges.  
   - This means that **any event in 𝓕 should be expressible as a limit of events depending on finitely many edges**.  

2. **Construction of a Sequence of Approximating Events**  
   - We need to construct a sequence **(𝐵ₙ)** where each **𝐵ₙ** depends on a finite number of edges and **converges to 𝐴 in probability**.  
   - Define **𝐵ₙ** as the event that depends only on the states of edges within a finite box **Λₙ** of radius **n** around the origin.  

3. **Limit Argument Using Increasing Events**  
   - Since percolation configurations **𝜔** are defined on an **infinite** graph, we approximate them by considering restrictions to finite subgraphs **Λₙ**.  
   - For large enough **n**, the probability measure of **𝐵ₙ** should be close to that of **𝐴**, ensuring:
     \[
     \lim_{n \to \infty} 𝒫ₚ[𝐵ₙ] = 𝒫ₚ[𝐴].
     \]
   - This follows from the **definition of the product measure 𝒫ₚ** and the fact that **local events stabilize in probability as n increases**.  

4. **Conclusion**  
   - The construction ensures that **any measurable event 𝐴 can be well approximated by events depending on finitely many edges**.  
   - This result justifies why percolation theory often **restricts focus to finite subgraphs** without loss of generality.

---

### **Key Takeaways**  
✅ **𝓕 is generated by finite-edge events, allowing approximation**  
✅ **Probability measure 𝒫ₚ allows limiting arguments**  
✅ **Events stabilize as n increases, ensuring convergence**

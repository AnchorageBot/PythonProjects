**Key Concepts:**
1. Hypercubic Lattice: d-dimensional grid of points with integer coordinates
2. Square Lattice: Special case of hypercubic lattice when d=2
3. Bernoulli Percolation: Random process where edges are independently open with probability p
4. Self-avoiding Path: Path where no vertex is visited more than once
5. Infinite Cluster: Connected component containing infinitely many vertices
6. Critical Point: Threshold probability where infinite clusters first appear
7. Dual Graph: Graph constructed by placing vertices in faces of original graph
8. Increasing Event: Event more likely to occur as more edges become open

* **Œ∏(p):** Probability that the origin (or a specific vertex) belongs to an infinite cluster in Bernoulli percolation with parameter p
* **inf:** Infimum, if a set has a smallest element, then the smallest element is the infimum for the set
* **pc ‚à∂=inf{p‚àà[0,1]‚à∂Œ∏(p) >0}:** Critical probability; the infimum of p such that Œ∏(p) > 0 (may be denoted with d if dimension is important)

**Find:**
1. Prove that pc(‚Ñ§) = 1, where:
   * pc(‚Ñ§) is the critical probability for Bernoulli percolation on the one-dimensional integer lattice
   * This means showing that for any p < 1, there is zero probability of an infinite cluster

2. Prove that pc(‚Ñ§ √ó {0,...,n}) = 1, where:
   * This is a "strip" lattice: infinite in one direction but finite (height n+1) in the other
   * Need to show that percolation only occurs at p = 1 in this case too

**Solution:**
1. **For pc(‚Ñ§) = 1:**
   * Consider any vertex x ‚àà ‚Ñ§
   * For an infinite cluster to exist, we need an infinite path of open edges in at least one direction
   * Probability of k consecutive edges being open = p·µè
   * For an infinite path: lim(k‚Üí‚àû) p·µè = 0 for any p < 1
   * Therefore Œ∏(p) = 0 for all p < 1
   * Thus pc = inf{p ‚àà [0,1] : Œ∏(p) > 0} = 1

2. **For pc(‚Ñ§ √ó {0,...,n}):**
   * Similar argument, but need to consider all possible paths
   * Any infinite path must eventually proceed horizontally
   * Can bound the probability using:
     - Finite number of possible vertical movements (due to strip height)
     - Need infinitely many horizontal steps
   * Show that probability still goes to 0 for any p < 1

The progression from one-dimensional to strip lattices in Exercise 2 is particularly interesting as it helps build intuition for more complex percolation problems.

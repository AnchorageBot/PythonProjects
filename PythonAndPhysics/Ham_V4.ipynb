### Hamiltonian Overview

The **Hamiltonian** is a fundamental concept in physics and mathematics, particularly in the field of classical mechanics and quantum mechanics. It is named after **William Rowan Hamilton**, an Irish mathematician and physicist who developed the Hamiltonian formulation of mechanics in the 19th century.

#### Definition:
The Hamiltonian is a function that describes the total energy of a system, expressed in terms of its generalized coordinates and momenta. It typically takes the form:

\[ H(q, p) = T(p) + V(q) \]

where:
- \( H \) is the Hamiltonian,
- \( T \) is the kinetic energy,
- \( V \) is the potential energy,
- \( q \) represents generalized coordinates,
- \( p \) represents generalized momenta.

#### Applications:
1. **Classical Mechanics**: It provides an alternative formulation of Newtonian mechanics, making it easier to analyze complex systems.
2. **Quantum Mechanics**: In quantum theory, the Hamiltonian operator corresponds to the total energy of the system and plays a crucial role in the Schr√∂dinger equation.
3. **Statistical Mechanics**: It is used to derive thermodynamic properties from the underlying microstates of a system.
4. **Control Theory**: In optimization problems, the Hamiltonian can help find optimal control strategies.

The Hamiltonian's structure allows for deep insights into the conservation laws and symmetries of physical systems, making it a cornerstone of modern physics.



The Hamiltonian for a simple harmonic oscillator can be expressed in terms of its generalized coordinates and momenta. For a mass \( m \) attached to a spring with spring constant \( k \), the Hamiltonian \( H \) is given by:

\[
$H(q, p) = \frac{p^2}{2m} + \frac{1}{2} k q^2$
\]




### Glossary of Symbols and Terms

| Symbol        | Description                                            |
|---------------|--------------------------------------------------------|
| \( H \)       | Hamiltonian (total energy of the system)              |
| \( q \)       | Displacement from the equilibrium position (generalized coordinate) |
| \( p \)       | Momentum, defined as \( p = m \frac{dq}{dt} \)       |
| \( $\frac{p^2}{2m} \$) | Kinetic energy of the mass                        |
| \( \frac{1}{2} k q^2 \) | Potential energy stored in the spring           |
```

This Hamiltonian represents the total energy of the simple harmonic oscillator as the sum of its kinetic and potential energy.

Here is the quantum mechanical Hamiltonian for a simple harmonic oscillator:

### Hamiltonian:
\[
$\hat{H} = \frac{\hat{p}^2}{2m} + \frac{1}{2} m \omega^2 \hat{x}^2
$\]

### Explanation:
This Hamiltonian represents the total energy of a particle in a harmonic oscillator potential. The first term corresponds to the kinetic energy, and the second term represents the potential energy of the particle.

### Glossary of Symbols and Terms

| **Symbol**        | **Description**                           |
|-------------------|-------------------------------------------|
| \(\hat{H}\)       | Hamiltonian operator (total energy)       |
| \(\hat{p}\)       | Momentum operator                        |
| \(m\)             | Mass of the particle                     |
| \(\omega\)        | Angular frequency of the oscillator       |
| \(\hat{x}\)       | Position operator                        |
| \(\frac{\hat{p}^2}{2m}\) | Kinetic energy term               |
| \(\frac{1}{2} m \omega^2 \hat{x}^2\) | Potential energy term |




````python
# A Python script that demonstrates the application of Hamiltonian mechanics to a one-dimensional simple harmonic oscillator.
# It solves Hamilton's equations of motion and visualizes the results, including position, momentum, and total energy over time.
# Additionally, it plots the phase space trajectory (position vs. momentum) to show the dynamics of the system.

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters of the system
m = 1.0      # Mass of the particle (kg)
omega = 1.0  # Angular frequency (rad/s)

# Hamilton's equations of motion for the simple harmonic oscillator
def hamiltonian_eqns(t, z, m, omega):
    x, p = z
    dxdt = p / m  # dx/dt = p / m
    dpdt = -m * omega**2 * x  # dp/dt = -m * omega^2 * x (Hooke's law)
    return [dxdt, dpdt]

# Total energy of the system (Hamiltonian)
def total_energy(x, p, m, omega):
    kinetic_energy = p**2 / (2 * m)
    potential_energy = 0.5 * m * omega**2 * x**2
    return kinetic_energy + potential_energy

# Initial conditions: position x0 and momentum p0
x0 = 1.0  # Initial position (m)
p0 = 0.0  # Initial momentum (kg m/s)

# Time array for the simulation
t_span = (0, 10)  # Time span from 0 to 10 seconds
t_eval = np.linspace(*t_span, 1000)  # Time points for evaluation

# Solve Hamilton's equations using SciPy's solve_ivp
sol = solve_ivp(hamiltonian_eqns, t_span, [x0, p0], args=(m, omega), t_eval=t_eval)

# Extract position and momentum from the solution
x = sol.y[0]  # Position over time
p = sol.y[1]  # Momentum over time

# Calculate total energy at each time step
energy = total_energy(x, p, m, omega)

# Plot the results
fig, ax = plt.subplots(3, 1, figsize=(8, 10))

# Plot position vs time
ax[0].plot(sol.t, x, label='Position (x)')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Position (m)')
ax[0].set_title('Position vs Time')
ax[0].grid(True)
ax[0].legend()

# Plot momentum vs time
ax[1].plot(sol.t, p, label='Momentum (p)', color='orange')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Momentum (kg m/s)')
ax[1].set_title('Momentum vs Time')
ax[1].grid(True)
ax[1].legend()

# Plot total energy vs time
ax[2].plot(sol.t, energy, label='Total Energy', color='green')
ax[2].set_xlabel('Time (s)')
ax[2].set_ylabel('Energy (J)')
ax[2].set_title('Total Energy vs Time (Conservation of Energy)')
ax[2].grid(True)
ax[2].legend()

plt.tight_layout()
plt.show()

# Phase space plot (Position vs Momentum)
plt.figure(figsize=(6, 6))
plt.plot(x, p, label='Phase Space (x vs p)', color='purple')
plt.xlabel('Position (x) [m]')
plt.ylabel('Momentum (p) [kg m/s]')
plt.title('Phase Space Trajectory')
plt.grid(True)
plt.legend()
plt.show()

````

Given:

* $y = \frac{x^2}{3} + \frac{1}{x}$

Find:

* That it is a solution of $xy' + y = x^2$ on $(0, {\infty})$ and $(-{\infty}, 0)$

Solution

* Take the derivative of the first equation $y = \frac{x^2}{3} + \frac{1}{x}$

  * Use the power rule $\frac{d}{dx}(x^n) = nx^{n-1}$

  * $\frac{d}{dx}(\frac{x^2}{3}) = \frac{1}{3}\frac{d}{dx}(x^2) = \frac{1}{3} \cdot 2x$

  * $\frac{d}{dx}(\frac{1}{x}) = \frac{d}{dx}(x^{-1}) = -x^{-2} = -\frac{1}{x^{2}}\$

  * y' = $\frac{2x}{3} - \frac{1}{x^{2}}\$

* Substitute

  * $y = \frac{x^2}{3} + \frac{1}{x}$

  * y' = $\frac{2x}{3} - \frac{1}{x^{2}}\$

* Into

  * $xy' + y = x^2$

* Yielding

  * x($\frac{2x}{3} - \frac{1}{x^{2}}) + \frac{x^2}{3} + \frac{1}{x} = x^2$

* Left Hand Side (LHS), distribute x:

  * $\frac{2x^2}{3} - \frac{1}{x} + \frac{x^2}{3} + \frac{1}{x} = x^2$

* LHS, combine like terms:

  * $\frac{2x^2}{3} + \frac{x^2}{3} - \frac{1}{x} + \frac{1}{x} = x^2$

  * $\frac{3x^2}{3} + 0 = x^2$

* LHS, simplify:

  * $x^2 = x^2$

```python
"""
Abstract:
This script solves and plots the solution to the first-order linear differential equation
xy' + y = x^2 on the intervals (0, ∞) and (-∞, 0). The equation is first rearranged into
the form y' = (x^2 - y) / x, which is then solved numerically using SciPy's odeint function.
The solution is plotted for both positive and negative x values, demonstrating the behavior
of the solution as x approaches positive and negative infinity. The script uses NumPy for
numerical computations, SciPy for solving the ODE, and Matplotlib for visualization.
"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def ode(y, x):
    """
    Define the ordinary differential equation xy' + y = x^2 in the form y' = f(x, y).

    Args:
    y (float): The dependent variable
    x (float): The independent variable

    Returns:
    float: The derivative dy/dx
    """
    return (x**2 - y) / x

# Generate x values, avoiding x=0 as it's a singular point
x_positive = np.linspace(0.01, 10, 1000)  # 1000 points between 0.01 and 10
x_negative = np.linspace(-10, -0.01, 1000)  # 1000 points between -10 and -0.01

# Solve ODE for positive x
# Initial condition y(0.01) ≈ 0 is assumed
y_positive = odeint(ode, 0, x_positive)

# Solve ODE for negative x
# Initial condition y(-0.01) ≈ 0 is assumed
y_negative = odeint(ode, 0, x_negative)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot solutions
plt.plot(x_positive, y_positive, 'b-', label='x > 0')
plt.plot(x_negative, y_negative, 'r-', label='x < 0')

# Set title and labels
plt.title("Solution to xy' + y = x^2")
plt.xlabel('x')
plt.ylabel('y')

# Add legend and grid
plt.legend()
plt.grid(True)

# Add asymptote at y=0
plt.axhline(y=0, color='k', linestyle='--', label='y = 0 asymptote')

# Set axis limits for better visualization
plt.xlim(-10, 10)
plt.ylim(-10, 40)

# Display the plot
plt.show()
```

A gambler with a stake of  50 dollars tries his luck with a series of 1 dollar bets until they either hit 100 dollars or go bust



Import the following libary tools

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

Draft up a function which uses a constant bet size

```python
def gambler():
  stake = 50
  goal = 100
  simulation_count =[]

  while stake in range(1,goal):
    bet_size = 1
    win_loss = random.randrange(-1,2,step=2)
    stake += bet_size * win_loss
    simulation_count.append(stake)

  return simulation_count

  #print(win_loss)
```

Plot out the function

```python
plt.plot(gambler())
plt.yticks(np.arange(-20,120,10))
plt.axhline(y=0,color='r',linestyle='dashdot')
plt.axhline(y=100,color='blue', linestyle='solid')
plt.xlabel('Number of bets')
plt.ylabel('Winnings')
plt.title('Gambling Simulation')
```

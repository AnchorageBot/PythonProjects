The exponential distribution is a continuous probability distribution that describes the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate. It's often used to model the time until the next event occurs, such as the time until a machine part fails, or the time between arrivals at a service queue.

Key properties of the exponential distribution:
1. Memoryless property: The future lifetime of an exponentially distributed random variable is independent of its current age.
2. Constant hazard rate: The rate at which events occur remains constant over time.
3. The mean and standard deviation are equal to 1/λ, where λ is the rate parameter.

Glossary of Terms and Symbols:

| Term/Symbol | Description |
|-------------|-------------|
| λ (lambda)  | Rate parameter of the exponential distribution. Represents the average number of events per unit time. |
| PDF         | Probability Density Function. For exponential distribution: f(x) = λe^(-λx) for x ≥ 0 |
| CDF         | Cumulative Distribution Function. For exponential distribution: F(x) = 1 - e^(-λx) for x ≥ 0 |
| E(X)        | Expected value (mean) of the distribution. For exponential: E(X) = 1/λ |
| Var(X)      | Variance of the distribution. For exponential: Var(X) = 1/λ^2 |
| σ (sigma)   | Standard deviation. For exponential: σ = 1/λ |
| Memoryless property | P(X > s + t | X > s) = P(X > t) for all s, t ≥ 0 |
| Hazard rate | h(x) = λ (constant for exponential distribution) |


Python script with three examples of exponential distributions:

```python
# Abstract:
# This script demonstrates the exponential distribution using three different
# rate parameters (λ). It plots the probability density function (PDF) and
# cumulative distribution function (CDF) for each case, and calculates key
# statistics. The script uses NumPy for numerical operations, Matplotlib for
# plotting, and SciPy for statistical functions.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set up the main figure with 2 rows and 3 columns of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Define the range of x values
x = np.linspace(0, 5, 1000)

# Define three different rate parameters (λ)
lambdas = [0.5, 1, 2]

for i, lambda_param in enumerate(lambdas):
    # Calculate the PDF and CDF
    pdf = stats.expon.pdf(x, scale=1/lambda_param)
    cdf = stats.expon.cdf(x, scale=1/lambda_param)
    
    # Plot PDF
    axs[0, i].plot(x, pdf, 'r-', lw=2)
    axs[0, i].set_title(f'Exponential PDF (λ={lambda_param})')
    axs[0, i].set_xlabel('x')
    axs[0, i].set_ylabel('Probability Density')
    
    # Plot CDF
    axs[1, i].plot(x, cdf, 'b-', lw=2)
    axs[1, i].set_title(f'Exponential CDF (λ={lambda_param})')
    axs[1, i].set_xlabel('x')
    axs[1, i].set_ylabel('Cumulative Probability')
    
    # Calculate and display statistics
    mean = 1 / lambda_param
    variance = 1 / (lambda_param ** 2)
    std_dev = np.sqrt(variance)
    
    stats_text = f'Mean: {mean:.2f}\nVariance: {variance:.2f}\nStd Dev: {std_dev:.2f}'
    axs[0, i].text(0.95, 0.95, stats_text, transform=axs[0, i].transAxes,
                   verticalalignment='top', horizontalalignment='right',
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()

# Generate random samples from each distribution
samples = 1000
for lambda_param in lambdas:
    random_samples = np.random.exponential(scale=1/lambda_param, size=samples)
    print(f"\nStatistics for λ={lambda_param}:")
    print(f"Sample Mean: {np.mean(random_samples):.4f}")
    print(f"Sample Variance: {np.var(random_samples):.4f}")
    print(f"Sample Standard Deviation: {np.std(random_samples):.4f}")
```

This script does the following:

1. It creates a figure with 6 subplots: 3 for PDFs and 3 for CDFs.
2. It defines three different rate parameters (λ): 0.5, 1, and 2.
3. For each λ, it plots the PDF and CDF of the exponential distribution.
4. It calculates and displays the theoretical mean, variance, and standard deviation for each distribution.
5. It generates random samples from each distribution and prints sample statistics.

The script demonstrates how the shape of the exponential distribution changes with different rate parameters. A higher λ results in a steeper decline in the PDF and a faster approach to 1 in the CDF, indicating that events are more likely to occur sooner.

The sample statistics printed at the end allow for comparison between the theoretical values and those obtained from random sampling, illustrating the law of large numbers in action.

````python
# Abstract:
# This script demonstrates the exponential distribution using three different
# rate parameters (λ). It plots the probability density function (PDF) and
# cumulative distribution function (CDF) for each case, and calculates key
# statistics. The script uses NumPy for numerical operations, Matplotlib for
# plotting, and SciPy for statistical functions.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set up the main figure with 2 rows and 3 columns of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Define the range of x values
x = np.linspace(0, 5, 1000)

# Define three different rate parameters (λ)
lambdas = [0.5, 1, 2]

for i, lambda_param in enumerate(lambdas):
    # Calculate the PDF and CDF
    pdf = stats.expon.pdf(x, scale=1/lambda_param)
    cdf = stats.expon.cdf(x, scale=1/lambda_param)

    # Plot PDF
    axs[0, i].plot(x, pdf, 'r-', lw=2)
    axs[0, i].set_title(f'Exponential PDF (λ={lambda_param})')
    axs[0, i].set_xlabel('x')
    axs[0, i].set_ylabel('Probability Density')

    # Plot CDF
    axs[1, i].plot(x, cdf, 'b-', lw=2)
    axs[1, i].set_title(f'Exponential CDF (λ={lambda_param})')
    axs[1, i].set_xlabel('x')
    axs[1, i].set_ylabel('Cumulative Probability')

    # Calculate and display statistics
    mean = 1 / lambda_param
    variance = 1 / (lambda_param ** 2)
    std_dev = np.sqrt(variance)

    stats_text = f'Mean: {mean:.2f}\nVariance: {variance:.2f}\nStd Dev: {std_dev:.2f}'
    axs[0, i].text(0.95, 0.95, stats_text, transform=axs[0, i].transAxes,
                   verticalalignment='top', horizontalalignment='right',
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()

# Generate random samples from each distribution and create a table of statistics
samples = 1000
print("\nSample Statistics:")
print(f"{'λ':^10}{'Mean':^15}{'Variance':^15}{'Std Dev':^15}")
print("-" * 55)

for lambda_param in lambdas:
    random_samples = np.random.exponential(scale=1/lambda_param, size=samples)
    sample_mean = np.mean(random_samples)
    sample_var = np.var(random_samples)
    sample_std = np.std(random_samples)
    print(f"{lambda_param:^10.2f}{sample_mean:^15.4f}{sample_var:^15.4f}{sample_std:^15.4f}")

# Print theoretical values for comparison
print("\nTheoretical Values:")
print(f"{'λ':^10}{'Mean':^15}{'Variance':^15}{'Std Dev':^15}")
print("-" * 55)

for lambda_param in lambdas:
    theo_mean = 1 / lambda_param
    theo_var = 1 / (lambda_param ** 2)
    theo_std = np.sqrt(theo_var)
    print(f"{lambda_param:^10.2f}{theo_mean:^15.4f}{theo_var:^15.4f}{theo_std:^15.4f}")
````

Let's review some rules of algebra

* 1/$x^n$ = $x^{-n}$
* $\sqrt{x}$ = x$^{1/2}$
* $x^0$ = 1
* $x^m$$x^n$ = $x^{m+n}$
* $x^m$/$x^n$ = $x^{m-n}$
* ($x^m$)$^n$ = $x^{mn}$
* (ax)$^m$ = $a^m$$x^m$
*  $\sqrt{ax}$ = (ax)$^{1/2}$ = $a^{1/2}$$x^{1/2}$
* A function assigns each element of a set (a mathematical collection of things) X to exactly one element of a set Y.

Remember that a derivative:
* Describes the rate of change
* The operation itself can denoted by a variety of symbols
> * f'(t) = df/dt = d/dt {f(t)}
* Can be solved by selectively applying rules
> * d/dx (constant c) = 0
> * d/dx (a$x^b$) = bax$^{b-1}$
> * d/dx ($y_1$ + $y_2$ + ... + $y_n$) = d$y_1$/dx + d$y_2$/dx + ... + d$y_n$/dx
> * Chain rule
> > * Finds the derivative of f(u(x)) with respect to x
> > * f ' (u(x)) = $\frac{df}{dx}$ = $\frac{df}{du}$ $\frac{du}{dx}$
> * Product rule
> > * Finds the derivative of the product of f(x) and g(x) with respect to x
> > * f ' (fg) = $\frac{d}{dx}$ (fg) = g($\frac{df}{dx}$) + f($\frac{dg}{dx}$)
> * Quotient rule
> > * Finds the derivative of the quotient of f(x) and g(x) with respect to x
> > * f ' ($\frac{f}{g}$) = $\frac{d}{dx}$ ($\frac{f}{g}$) = [g($\frac{df}{dx}$) - f($\frac{df}{dx}$)]/g$^2$

Let's take the derivative of f(x) = (3x$^2$ - 7)(2X$^3$ - 6)
* We will use the product rule
> * f ' (fg) = $\frac{d}{dx}$ (fg) = g($\frac{df}{dx}$) + f($\frac{dg}{dx}$)
* This gives us
> * f(x) = 3x$^4$ - 7
> * g(x) = 2x$^3$ - 6
> * f ' (fg) = $\frac{d}{dx}$ (fg) = g($\frac{df}{dx}$) + f($\frac{dg}{dx}$)
> > * g($\frac{df}{dx}$) = (2x$^3$ - 6)[$\frac{d}{dx}$ 3x$^4$ - 7] = (2x$^3$ - 6) [12x$^3$]
> > * f($\frac{dg}{dx}$) = (3x$^4$ - 7)[$\frac{d}{dx}$ 2x$^3$ - 6] = (3x$^4$ - 7) [6x$^2$]
> * f ' (fg) = $\frac{d}{dx}$ (fg) = (2x$^3$ - 6) [12x$^3$] + (3x$^4$ - 7) [6x$^2$]

Let's graph the function of f(x) = (3x$^2$ - 7)(2X$^3$ - 6) with matplotlib:

```python
import matplotlib.pyplot as plt

xmin = -4
xmax = 4
ymin = -1400
ymax = 1400

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.title('f(x) = (3x^4 - 7)(2x^3 - 6)')
plt.xlabel('x')
plt.ylabel('y = (3x^4 - 7)(2x^3 - 6)')

for x in range(-5,5): # this graphs a line up and to the left and to the right
    y = (((3*x**4)-7)*(2*x**3 -6)) # function
    plt.plot([x],[y], 'ro') # r = red , o = marker, this makes the red dots

plt.show()
```

Let's calculate the derivative of f(x) = (3x$^4$ - 7)(2X$^3$ - 6) with respect to x, and check our answer by integrating it, using SymPy. Then let's graph it using matplotlib

```python
import matplotlib.pyplot as plt

import sympy as sy
from sympy import *

sy.init_printing()
init_printing(use_unicode = True)

#x = symbols('x')
#f,g = symbols('f g', cls = Function)
#(f(x)*g(x)).diff(x)
#print("\n")

x, y = symbols('x y')
y = ((3*x**4)-7)*((2*x**3)-6)
expr = y.diff(x)
expr
z = simplify(expr)
print("The derivative of the function f(x) = ((3*x**4)-7)*((2*x**3)-6) with respect to x is: f'(x) =", z)
print("Using latex format the derivative is, f'(x) =", sy.latex(sy.simplify(z)))
print("\n")

print('The plot of the derivative \n')
plot(z)
print("\n")

#print("The integral is the inverse of the derivative and can be used as a check (test) \n")
#w = integrate(z)
#print("Integrating the derivative yields: ", z)
#print("Using latex format integrating the derivative yields: ", sy.latex(sy.simplify(w)))
#print("\n")

#print('The plot of the integral \n')
#plot(w)
#print("\n")
```

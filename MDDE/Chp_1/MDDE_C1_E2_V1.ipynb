### Kinematics on a line

Given:

A particle moves along a horizontal line during the course of a time period. It's movement can be described by the following equation:

x = ϕ(t) = t<sup>2</sup> - t - 2

| Symbol | Meaning |
|--------|---------|
| $x$ | Horizontal displacement |
| $ϕ$ | Function relating time to displacement |
| $t$ | Time |

Find:

Sketch the graph of the function

```python
"""
Abstract:
This script models and visualizes the kinematics of a particle moving along a horizontal line.
The particle's position is described by the equation x = t² - t - 2, where x is the horizontal
displacement in meters and t is the elapsed time in seconds. The script provides functions
to calculate the displacement for given time values and to plot the displacement-time graph.
This visualization helps in understanding the particle's motion, including its starting position,
when it crosses the origin, and how its velocity changes over time.
"""

import numpy as np
import matplotlib.pyplot as plt

def calc_displacement(t):
    """
    Calculate the horizontal displacement of a particle along a line during a time period.

    The displacement is calculated using the equation: x = t² - t - 2

    Parameters:
    t (float or array-like): Time in seconds

    Returns:
    float or array-like: The horizontal displacement of the particle in meters (m)
    """
    return t**2 - t - 2

def plot_displacement(t_range):
    """
    Generate a graph of the displacement function.

    This function calculates the displacement for a given range of time values
    and plots the resulting displacement-time graph.

    Parameters:
    t_range (array-like): Range of time values to plot

    Returns:
    None
    """

    # Calculate displacement for each time value using the equation x = t² - t - 2
    x = calc_displacement(t_range)

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(t_range, x, label='Displacement')

    # Set title and labels
    plt.title('Horizontal Displacement vs Elapsed Time')
    plt.xlabel('Elapsed Time (sec)')
    plt.ylabel('Horizontal Displacement (meters)')

    # Add grid, legend, and axes
    plt.grid(True)
    plt.legend()
    plt.axhline(y=0, color='r', linestyle='--', label='x-axis')  # Add x-axis
    plt.axvline(x=0, color='r', linestyle='--', label='y-axis')  # Add y-axis

    # Display the plot
    plt.show()

def main():
    """
    Main function to execute the script.

    This function sets up the time range for analysis and calls the plotting function.
    """
    # Generate a range of time values from -2 to 4 seconds
    # The range is chosen to show the particle's motion before and after t=0
    t_range = np.linspace(-2, 4, 100)

    # Generate and display the displacement-time graph
    plot_displacement(t_range)

if __name__ == "__main__":
    main()
```

When choosing axes for a graph, we typically consider both convention and what best represents the data. Let's analyze this for our specific case of horizontal displacement versus elapsed time:

1. Convention:
   In physics and kinematics, it's common practice to plot time on the x-axis (horizontal) and position or displacement on the y-axis (vertical). This aligns with how we typically think about motion - progressing through time from left to right, with vertical changes representing changes in position.

2. Independent vs Dependent Variables:
   In general scientific plotting, the independent variable (the one you control or that naturally progresses) is usually placed on the x-axis, while the dependent variable (the one that changes as a result) is on the y-axis. In this case, time is the independent variable, and displacement depends on time.

3. Intuitive Reading:
   Plotting time on the x-axis allows for an intuitive left-to-right reading of the graph, which matches how we typically process information in many cultures.

4. Mathematical Function Representation:
   The equation x = ϕ(t) = t² - t - 2 is in the form y = f(x), where t is the input (like x) and x (displacement) is the output (like y). In mathematical function plotting, we typically put the input on the x-axis and the output on the y-axis.

Considering these points, the optimal choice would be:

- X-axis: Elapsed Time
- Y-axis: Horizontal Displacement

This choice:
- Follows conventions in physics and kinematics
- Correctly represents the independent (time) and dependent (displacement) variables
- Allows for intuitive left-to-right reading of the progression of motion
- Aligns with standard mathematical function representation

So, in this case, our current axis choice in the code is indeed optimal. It allows readers familiar with physics conventions to quickly understand and interpret the graph, while also accurately representing the mathematical and physical relationship between time and displacement.

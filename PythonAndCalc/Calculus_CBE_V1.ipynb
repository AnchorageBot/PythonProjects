The growth of a birch tree can be descibed by a function in mathematics.

A function assigns each element of a set (a mathematical collection of things) X to exactly one element of a set Y. For example, let's assign time data to the set X and tree radius data to the set Y. Let's arbitrarily use 3 years as our observation period and thus set X = {0,1,2,3}. Let's arbitrarily say that the tree's radius is measured each year and yields the set Y = {0,1,4,9}

Let's graph this function's output with matplotlib:

```python
import matplotlib.pyplot as plt

xmin = -1
xmax = 5
ymin = -1
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.title('birch tree growth')
plt.xlabel('years')
plt.ylabel('radius of the tree')

for x in range(4): # this graphs a line up and to the right
    y = x**2 # radius of the tree
    plt.plot([x],[y], 'ro') # r = red , o = marker, this makes the red dots

plt.show()
```

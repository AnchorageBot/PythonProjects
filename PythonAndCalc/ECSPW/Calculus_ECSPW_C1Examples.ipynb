Let's review some rules of algebra

* 1/$x^n$ = $x^{-n}$
* $\sqrt{x}$ = x$^{1/2}$
* $x^0$ = 1
* $x^m$$x^n$ = $x^{m+n}$
* $x^m$/$x^n$ = $x^{m-n}$
* ($x^m$)$^n$ = $x^{mn}$
* (ax)$^m$ = $a^m$$x^m$
*  $\sqrt{ax}$ = (ax)$^{1/2}$ = $a^{1/2}$$x^{1/2}$
* A function assigns each element of a set (a mathematical collection of things) X to exactly one element of a set Y.

Remember that a derivative:
* Describes the rate of change
* The operation itself can denoted by a variety of symbols
> * f'(t) = df/dt = d/dt {f(t)}
* Can be solved by using the formulas
> * d/dx (constant c) = 0
> * d/dx (a$x^b$) = bax$^{b-1}$
> * d/dx ($y_1$ + $y_2$ + ... + $y_n$) = d$y_1$/dx + d$y_2$/dx + ... + d$y_n$/dx

Let's take the derivative of f(x) = $x^2$ as a warm-up for the brain cell
* We will use the notation
> * f ' (x) = d/dx {f(x)}
* We will use the rule
> * d/dx (a$x^b$) = bax$^{b-1}$
* This gives us
> * d/dx ($x^2$) = 2x

Let's graph the derivative of f'(x) = d/dx ($x^2$) = 2x with matplotlib:

```python
import matplotlib.pyplot as plt

xmin = -1
xmax = 5
ymin = -1
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

plt.title('rate of change')
plt.xlabel('dx')
plt.ylabel('d/dx {f(x)} = 2x')

for x in range(4): # this graphs a line up and to the right
    y = 2*x # rate of change
    plt.plot([x],[y], 'ro') # r = red , o = marker, this makes the red dots

plt.show()
```

A derivative can be also described using words
> * derive [ equation ] with respect to [ variable ]
> * [ equation ] = x**2 = x squared
> * [ variable ] = x
> * derive [ x squared ] with respect to [ x ]
> * derive x squared with respect to x

Let's calculate and graph the derivative using Wolfram Alpha

Derivative_via_WA.png

Let's calculate the derivative of x squared with respect to x, and check our answer by integrating it, using SymPy. Then let's graph it using matplotlib

```python
import matplotlib.pyplot as plt

from sympy import *
x = symbols('x')
#init_printing(use_unicode = True)

x_deriv = diff(x**2,x)
print("The derivative of x**2 with respect to x is", x_deriv, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w = integrate(x_deriv)
print("The integral of 2x is", w, "\n")
print("\n")

print('The plot of the derivative \n')
plot(x)
print("\n")

print('The plot of the integral \n')
plot(w)
print("\n")
```

Chapter 1, Example 3

Let's take the derivative of f(x) = $6x^{2/3}$ with respect to x
* We will use the notation
> * f ' (x) = d/dx {f(x)}
* We will use the rule
> * d/dx (a$x^b$) = bax$^{b-1}$
* This gives us the resultant derivative
> * d/dx ($6x^{2/3}$) = (2/3)(6)$x^{2/3 - 1}$ = 4$x^{- 1/3}$
* Let's calculate the derivative, and check it by taking the integral, in SymPy:

```python
import matplotlib.pyplot as plt

from sympy import *
x = symbols('x')
#init_printing(use_unicode = True)

ex_3 = diff(6*x**(2/3),x)
print("The derivative of 6.0*x**0.666666666666667 with respect to x is", ex_3, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w3 = integrate(ex_3)
print("The integral of 4.0/x**0.333333333333333 is", w3, "\n")
print("Remember that 4.0/x**0.333333333333333 = 4*x** - 0.333333333333333")
print("\n")
```

Chapter 1, Example 4

Let's take the derivative of f(u) = 7/u = 7$u^{-1}$ with respect to u
* We will use rule
> * d/dx (a$x^b$) = bax$^{b-1}$
* We can restate the rule in the context of u
> * d/du (a$u^b$) = bau$^{b-1}$
* This gives us the resultant derivative
> * d/du (7/u) = d/du (7$u^{-1}$) = (-1)(7)$u^{-1-1}$ = - 7$u^{-2}$
* Let's calculate the derivative, and check it by taking the integral, in SymPy:

```python
import matplotlib.pyplot as plt

from sympy import *
u = symbols('u')
#init_printing(use_unicode = True)

ex_4 = diff(7*u**(-1),u)
print("The derivative of 7*u**(-1) with respect to u is", ex_4, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w4 = integrate(ex_4)
print("The integral of -7/u**2 is", w4, "\n")
print("\n")
```

Chapter 1, Example 5

Let's take the derivative of f(x) = $\sqrt{3x}$ with respect to x
* We will use the notation
> * f ' (x) = d/dx {f(x)}
* We will use the rules
> * $\sqrt{ax}$ = (ax)$^{1/2}$ = $a^{1/2}$$x^{1/2}$
> * d/dx (a$x^b$) = bax$^{b-1}$
* This gives us the resultant derivative
> * d/dx ($\sqrt{3x}$) = d/dx ($3^{1/2}$$x^{1/2}$)
> * d/dx ($3^{1/2}$$x^{1/2}$) = (1/2)($3^{1/2}$)($x^{1/2-1}$)
> * d/dx ($3^{1/2}$$x^{1/2}$) = (($3^{1/2}$)/2)($x^{-1/2}$)
* Let's calculate the derivative, and check it by taking the integral, in SymPy:

```python
import matplotlib.pyplot as plt

from sympy import *
x = symbols('x')
#init_printing(use_unicode = True)

ex_5 = diff(sqrt(3*x),x)
print("The derivative of sqrt(3*x) with respect to x is", ex_5, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w5 = integrate(ex_5)
print("The integral of sqrt(3)/(2*sqrt(x)) is", w5, "\n")
print("\n")
```

Chapter 1, Example 6

Let's take the derivative of f(x) = ($6x^3$ - 12x)  with respect to x
* We will use the notation
> * f ' (x) = d/dx {f(x)}
* We will use the rules
> * $x^0$ = 1
> * d/dx (a$x^b$) = bax$^{b-1}$
> * d/dx ($y_1$ + $y_2$ + ... + $y_n$) = d$y_1$/dx + d$y_2$/dx + ... + d$y_n$/dx
* This gives us the resultant derivative
> * f(x) = ($6x^3$ - 12x)
> * d/dx ($6x^3$ - 12x) = (3)(6)($x^{3 - 1}$) - (1)(12)(x$^{1-1}$)
> * d/dx ($6x^3$ - 12x) = 18$x^{2}$ - 12
* Let's calculate the derivative, and check it by taking the integral, in SymPy:

```python
import matplotlib.pyplot as plt

from sympy import *
x = symbols('x')
#init_printing(use_unicode = True)

ex_6 = diff((6*x**3 - 12*x),x)
print("The derivative of (6*x**3 - 12*x) with respect to x is", ex_6, "\n")

print("The integral is the inverse of the derivative and can be used as a check (test) \n")

w6 = integrate(ex_6)
print("The integral of (18*x**2 - 12) is", w6, "\n")
print("\n")
```

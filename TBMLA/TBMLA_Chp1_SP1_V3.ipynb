This script simulates rolling two six-sided dice 100 times, plots the distribution of the sums,

proper spacing, and the sum table has a clear separation between its title and horizontal numbers.
'''

import random
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

class Dice:
    """
    A class representing a pair of six-sided dice.
    """

    def __init__(self, sides=6):
        """
        Initialize the Dice object.

        Args:
            sides (int): Number of sides on each die. Defaults to 6.
        """
        self.sides = sides

    def roll(self):
        """
        Simulate rolling two dice and return their sum.

        Returns:
            int: The sum of two dice rolls.
        """
        return random.randint(1, self.sides) + random.randint(1, self.sides)

def simulate_rolls(dice, num_rolls):
    """
    Simulate rolling the dice a specified number of times.

    Args:
        dice (Dice): The Dice object to use for rolling.
        num_rolls (int): The number of rolls to simulate.

    Returns:
        Counter: A Counter object with the frequency of each sum.
    """
    return Counter(dice.roll() for _ in range(num_rolls))

def plot_results(roll_results):
    """
    Plot the distribution of dice roll sums.

    Args:
        roll_results (Counter): The frequency of each sum.
    """
    plt.figure(figsize=(10, 6))
    sums, frequencies = zip(*sorted(roll_results.items()))
    
    plt.bar(sums, frequencies)
    plt.title('Distribution of 100 Dice Rolls (Sum of Two Dice)')
    plt.xlabel('Sum of Dice')
    plt.ylabel('Frequency')
    plt.xticks(range(min(sums), max(sums) + 1))
    
    plt.tight_layout()
    plt.show()

def plot_sum_table():
    """
    Plot the sum table for two six-sided dice.
    """
    # Create the sum table
    table = np.zeros((6, 6), dtype=int)

        for j in range(6):

    # Create the plot

    ax = plt.gca()

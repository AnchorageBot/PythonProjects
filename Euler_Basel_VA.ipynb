While this isn't Euler's proof itself (which involved complex analysis and innovative manipulation of infinite series), it demonstrates numerically that ζ(2) = π²/6. The irrationality of ζ(2) follows from this relationship since π² is irrational (which follows from the irrationality of π).

```python
# The script help's to visualize and calculate approximations of ζ(2)

# Import required libraries
import numpy as np          # NumPy for efficient numerical operations
import matplotlib.pyplot as plt  # For plotting our results
from math import pi        # We'll need π for Euler's formula

def calculate_zeta_2_series(n_terms):
    """
    Calculate ζ(2) using the direct series sum up to n terms
    Returns both the partial sum and an array of partial sums
    """
    # Create a list of terms: [1/1², 1/2², 1/3², ..., 1/n²]
    terms = [1/(k**2) for k in range(1, n_terms + 1)]

    # Calculate cumulative sums of the terms using NumPy
    # This gives us [1/1², (1/1² + 1/2²), (1/1² + 1/2² + 1/3²), ...]
    partial_sums = np.cumsum(terms)

    # Return both the final sum and the array of partial sums
    return partial_sums[-1], partial_sums

def calculate_pi_squared_over_6():
    """Calculate ζ(2) using Euler's result: ζ(2) = π²/6"""
    # Simply compute π² and divide by 6
    return pi**2 / 6

# Set the number of terms we'll use in our series approximation
n_terms = 1000

# Calculate ζ(2) using both methods
series_sum, partial_sums = calculate_zeta_2_series(n_terms)  # Using series
euler_value = calculate_pi_squared_over_6()                  # Using π²/6

# Print the results with 10 decimal places
print(f"ζ(2) approximation using {n_terms} terms of the series: {series_sum:.10f}")
print(f"ζ(2) using Euler's π²/6: {euler_value:.10f}")
print(f"Absolute difference: {abs(series_sum - euler_value):.10f}")

# Create a visualization of the convergence
plt.figure(figsize=(12, 6))  # Set the size of our plot

# Plot the partial sums as a blue line
plt.plot(range(1, n_terms + 1), partial_sums, 'b-', label='Partial sums')

# Add a horizontal red dashed line for π²/6
plt.axhline(y=euler_value, color='r', linestyle='--', label='π²/6')

# Label the axes and add a title
plt.xlabel('Number of terms')
plt.ylabel('Sum value')
plt.title('Convergence of ζ(2) series to π²/6')

# Add grid lines for better readability
plt.grid(True)

# Add a legend to explain the lines
plt.legend()

# Use logarithmic scale for x-axis to better show early convergence
plt.xscale('log')

# Display the plot
plt.show()

# Print a detailed convergence table for the first 10 terms
print("\nConvergence of first 10 terms:")
for n in range(1, 11):
    # Calculate partial sum for n terms
    partial_sum = sum(1/(k**2) for k in range(1, n + 1))

    # Calculate relative error as a percentage
    error = abs(partial_sum - euler_value) / euler_value * 100

    # Print formatted results
    print(f"n = {n:2d}: sum = {partial_sum:.10f}, relative error = {error:.6f}%")
```

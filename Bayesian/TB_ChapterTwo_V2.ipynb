Let's use a script to draw Venn Diagrams for the complement A, B
* Complement definition = collection of elements not in the set
* Complement symbology = B' = $B^c$



```python
# import libraries
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline

# setup drawing area
figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))

# Draw circle A
v1 = venn2(subsets=(3, 3, 1), ax=ax1)
c1 = venn2_circles(subsets=(3, 3, 1), ax=ax1)

for area in ['01', '10', '11']: # use for loops to go through each area
    color = 'skyblue' if area != '01' else 'white'
    v1.get_patch_by_id(area).set_color(color)
    v1.get_patch_by_id(area).set_alpha(1)
    txt = v1.get_label_by_id(area)
    if txt: txt.set_text('')

ax1.set_axis_on()
ax1.set_facecolor('white')
ax1.set_title("A, B'", fontsize=20)
ymin, ymax = ax1.get_ylim()
ax1.set_ylim(ymin - 0.1, ymax)

# Draw Circle B
v2 = venn2(subsets=(3, 3, 1), ax=ax2)
c2 = venn2_circles(subsets=(3, 3, 1), ax=ax2)

for area in ['01', '10', '11']:
    color = 'skyblue' if area != '10' else 'white'
    v2.get_patch_by_id(area).set_color(color)
    v2.get_patch_by_id(area).set_alpha(1)
    txt = v2.get_label_by_id(area)
    if txt: txt.set_text('')

ax2.set_axis_on()
ax2.set_facecolor('white')
ax2.set_title("B, A'", fontsize=20)
ymin, ymax = ax2.get_ylim()
ax2.set_ylim(ymin - 0.1, ymax)
```

Let's use a script to draw Venn Diagrams for the intersection A, B

* A AND B, A NAND B
* Intersection definition = set of elements in both A and B
* Intersection symbology = $A\bigcap B$

```python
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline

figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))

# A AND B
v1 = venn2(subsets=(3, 3, 1), ax=ax1)
c1 = venn2_circles(subsets=(3, 3, 1), ax=ax1)

for area in ['01', '10', '11']:
    color = 'skyblue' if area == '11' else 'white'
    v1.get_patch_by_id(area).set_color(color)
    v1.get_patch_by_id(area).set_alpha(1)
    txt = v1.get_label_by_id(area)
    if txt: txt.set_text('')

ax1.set_axis_on()
ax1.set_facecolor('white')
ax1.set_title("A AND B", fontsize=20)
ymin, ymax = ax1.get_ylim()
ax1.set_ylim(ymin - 0.1, ymax)

#  A NAND B
v2 = venn2(subsets=(3,3,1))
c2 = venn2_circles(subsets=(3,3,1))

for area in ['01', '10', '11']:
    color = "skyblue" if area != '11' else "white"
    v2.get_patch_by_id(area).set_color(color)
    v2.get_patch_by_id(area).set_alpha(1)
    txt = v2.get_label_by_id(area)
    if txt: txt.set_text('')

ax2.set_axis_on()
ax2.set_facecolor('skyblue')
ax2.set_title("A NAND B", fontsize=20)
ymin, ymax = ax2.get_ylim()
ax2.set_ylim(ymin - 0.1, ymax)

plt.show()
```

Let's use a script to draw Venn Diagrams for the union A, B
* A OR B, A NOR B
* Union definition = set of elements in A, or B, or both.
* Union symbology =  $A\bigcup B$

```python
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline

figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))

# A OR B
v1 = venn2(subsets=(3, 3, 1), ax=ax1)
c1 = venn2_circles(subsets=(3, 3, 1), ax=ax1)

for area in ['01', '10', '11']:
    v1.get_patch_by_id(area).set_color('skyblue')
    v1.get_patch_by_id(area).set_alpha(1)
    txt = v1.get_label_by_id(area)
    if txt: txt.set_text('')

ax1.set_axis_on()
ax1.set_facecolor('white')
ax1.set_title("A OR B", fontsize=20)
ymin, ymax = ax1.get_ylim()
ax1.set_ylim(ymin - 0.1, ymax)

# A NOR B
v2 = venn2(subsets=(3, 3, 1), ax=ax2)
c2 = venn2_circles(subsets=(3, 3, 1), ax=ax2)

for area in ['01', '10', '11']:
    v2.get_patch_by_id(area).set_color('white')
    v2.get_patch_by_id(area).set_alpha(1)
    txt = v2.get_label_by_id(area)
    if txt: txt.set_text('')

ax2.set_axis_on()
ax2.set_facecolor('skyblue')
ax2.set_title("A NOR B", fontsize=20)
ymin, ymax = ax2.get_ylim()
ax2.set_ylim(ymin - 0.1, ymax)

plt.show()
```

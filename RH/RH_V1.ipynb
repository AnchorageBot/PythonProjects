```python
"""
Abstract:
This script defines a limited, discrete toy zeta function that mimics some properties
of the Riemann zeta function. The function is designed to be zero for negative even
integers and at specific points on the critical line (Re(s) = 1/2). This toy function
is not meant to replicate the full complexity of the Riemann zeta function but rather
to demonstrate a simplified version with specific zero points.

Key properties:
1. Returns 0 for negative even integers.
2. Returns 0 for s = 0.5 + i, 0.5 + 2i, and 0.5 + 3i.
3. Returns 1 for all other inputs.
"""

import numpy as np
import matplotlib.pyplot as plt

def toy_zeta(s):
    """
    A toy zeta function that is zero for negative even integers and
    at specific points on the critical line.

    Args:
    s (int, float, or complex): The input value for the function.

    Returns:
    int: 0 if s is a negative even integer or one of the specified points on the critical line,
         1 otherwise.
    """
    if isinstance(s, (int, float, complex)):
        # Check if s is a negative even integer
        if isinstance(s, int) and s < 0 and s % 2 == 0:
            return 0

        # Check if s is on the critical line (Re(s) = 1/2) and has specific imaginary parts
        if np.isclose(s.real, 0.5) and isinstance(s, complex) and s.imag in [1, 2, 3]:
            return 0

    # For all other inputs, return a non-zero value
    return 1

# Test the function with various inputs
test_values = [-4, -2, 0, 1, 0.5, 0.5+1j, 0.5+2j, 0.5+3j, 0.5+4j, 1+1j]

# Iterate through test values and print results
for s in test_values:
    result = toy_zeta(s)
    print(f"toy_zeta({s}) = {result}")

# Create a graph to visualize the function
def plot_toy_zeta():
    # Create a grid of points
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j*Y

    # Compute the toy zeta function for each point
    W = np.frompyfunc(toy_zeta, 1, 1)(Z).astype(float)

    # Create the plot
    plt.figure(figsize=(12, 8))
    plt.pcolormesh(X, Y, W, cmap='viridis', shading='auto')
    plt.colorbar(label='toy_zeta(s)')
    plt.title('Toy Zeta Function')
    plt.xlabel('Re(s)')
    plt.ylabel('Im(s)')
    plt.axvline(x=0.5, color='r', linestyle='--', label='Critical Line')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the plotting function
plot_toy_zeta()
```

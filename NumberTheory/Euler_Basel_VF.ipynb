## Glossary of Symbols and Terms

| Symbol/Term             | Meaning                                                                                                                              |
| ----------------------- | ------------------------------------------------------------------------------------------------------------------------------------ |
| Basel Problem           | The problem of finding the sum of the reciprocals of the squares of the positive integers (Σ 1/k² from k=1 to infinity).                  |
| Weierstrass Factorization Theorem | A theorem that expresses an entire function as an infinite product. In this case, it's used for the sine function.             |
| L'Hôpital's Rule        | A rule for evaluating limits of indeterminate forms (like 0/0 or ∞/∞) by taking the derivatives of the numerator and denominator.       |
| Σ                     | Summation (the sum of a series).                                                                                                     |
| Π                     | Product (the product of a sequence of terms).                                                                                           |
| *x*                     | A variable.                                                                                                                         |
| *k*                     | An integer index for summation/product.                                                                                              |
| sin(π*x*)              | The sine function of π*x.                                                                                                            |
| cos(π*x*)              | The cosine function of π*x.                                                                                                           |
| ln(*x*)                 | The natural logarithm of *x*.                                                                                                         |
| lim (x→0)               | The limit as *x* approaches 0.                                                                                                       |
| k²                    | k squared (k*k)                                                                                                                        |
| π                     | Pi (approximately 3.14159)                                                                                                          |

## Equations and Steps

**Equation 1:**

```
sin(πx) = πx * Π (1 - x²/k²)  (product from k=1 to infinity)
```

*   This is the Weierstrass factorization theorem applied to sin(πx).

**Equation 2:**

```
ln(sin(πx)) = ln(πx) + Σ ln(1 - x²/k²)   (sum from k=1 to infinity)
```

*   **Step from Equation 1 to Equation 2:** The natural logarithm (ln) is applied to both sides.

**Equation 3:**

```
(π * cos(πx) / sin(πx)) = (1/x) - 2x * Σ (1 / (k² - x²))   (sum from k=1 to infinity)
```

*   **Step from Equation 2 to Equation 3:** Equation 2 is differentiated with respect to *x*.

**Equation 4:**

```
(π * cos(πx) / sin(πx)) - (1/x) = -2x * Σ (1 / (k² - x²))   (sum from k=1 to infinity)
```

*   **Step from Equation 3 to Equation 4:** The term (1/x) is subtracted from both sides.

**Equation 5:**

```
(sin(πx) - πx * cos(πx)) / (2x² * sin(πx)) = Σ (1 / (k² - x²))   (sum from k=1 to infinity)
```

*   **Step from Equation 4 to Equation 5:** Both sides are divided by -2x and the left side is rewritten by combining fractions.

**Equation 6:**

```
(sin(πx) - πx * cos(πx)) / (2x² * sin(πx)) = Σ (1 / (k² - x²))   (sum from k=1 to infinity); k² - x² ≠ 0
```

*   This is a restatement of equation 5.

**Equation 7:**

```
Σ (1 / k²) = 1 + (1/2²) + (1/3²) + (1/4²) + ...   (sum from k=1 to infinity)
```

*   **Step from Equation 6 to Equation 7:** This considers the limit of the *right-hand side* of Equation 6 as x approaches 0.

**Equation 8:**

```
lim (x→0) (sin(πx) - πx * cos(πx)) / (2x² * sin(πx)) → 0/0
```
